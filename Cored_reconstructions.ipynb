# gamma_c = 0.4 at Gs

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants and parameters
rc = 1
rs = 18.6  # Scale radius in kpc
rho0 = 0.383  # Local DM density in GeV/cm^3
r0 = 8.2  # Sun position in kpc
rho_sat = 3.17*10**11
gamma_sp = 2.25

gamma_c = 0.4

# Function to calculate the NFW halo density profile
def nfw_halo(r, rs):
    x = r / rs
    return rhos / (x * (1 + x)**2)



# Function to calculate the DM mass density of the Milky Way
def mw_core_GS(r,rc,gamma_c ,Rs, Rsat, Rsp,gamma_sp):
    density = np.zeros_like(r)

    # Condition 1: r < 2RS
    mask_condition_1 = r < 2 * Rs
    density[mask_condition_1] = 0

    # Condition 2: 2RS ≤ r < Rsat
    mask_condition_2 = (r >= 2 * Rs) & (r < Rsat)
    density[mask_condition_2] = rho_sat * (r[mask_condition_2] / Rsat)**(-0.5)

    # Condition 3: Rsat ≤ r < Rsp
    mask_condition_3 = (r >= Rsat) & (r < Rsp)
    density[mask_condition_3] = (nfw_halo(rc, rs)*(Rsp/rc)**(-gamma_c))*(r[mask_condition_3]/ Rsp)**(-gamma_sp)

    # Condition 4: r ≥ Rsp
    mask_condition_4 = (r >= Rsp) & (r < rc)
    density[mask_condition_4] = nfw_halo(rc, rs)*(r[mask_condition_4]/rc)**(-gamma_c)

    mask_condition_5 = r >= rc
    density[mask_condition_5] = nfw_halo(r[mask_condition_5], rs)

    return density
# Generate radii from 0.01 to 100 times the scale radius
r = np.logspace(-9, 0, 100)  # Adjusted range to start from a positive value

rhos = (rho0 * ((r0 / rs) * (1 + r0 / rs)**(2)))

# Parameters for the Milky Way density profile
Rs = 3.8*10**(-11)
Rsp = 0.34*10**(-3)
Rsat = (((rhos / rho_sat) * (rs / rc) * ((1 + (rc / rs))**(-2))*((rc/Rsp)**(gamma_c)))**(1 / gamma_sp)) * Rsp

# Calculate the density profile for the Milky Way
core_mw_GS = mw_core_GS(r,rc,gamma_c, Rs, Rsat, Rsp, gamma_sp)

# Plot the Milky Way density profile
plt.plot(r, core_mw_GS, label='GS', color='blue')
plt.xlabel('Radius (r in kpc)')
plt.ylabel('Density (rho in GeV/cm^3)')
plt.title('Milky Way DM Density Profile')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.legend()
plt.xlim([10**(-9), 1])  # Set the desired x-axis limits
plt.show()
```
