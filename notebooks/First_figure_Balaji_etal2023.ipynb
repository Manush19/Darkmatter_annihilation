```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import sys,json,os
sys.path.append('../tools')
import profiles as tp
from profiles import Constants as pc
from profiles import MWDensity as pm
```

```python
pm1 = pm(halo_types = ['NFW','Core'])
```

```python
fig,axs = plt.subplots(1,2,figsize = (12,4))
ax1,ax2 = axs

ax1.plot(pm1.r, pm1.density['NFW']['GS'], c = 'b', label = 'GS')
ax1.plot(pm1.r, pm1.density['NFW']['no spike'], c = 'k', label = 'no spike')
ax1.plot(pm1.r, pm1.density['NFW']['NA'], c = 'm', label = 'NA')
ax1.plot(pm1.r, pm1.density['NFW']['BM'], c = 'grey', label = 'BM')
ax1.plot(pm1.r, pm1.density['NFW']['SH-'], c = 'brown', label = r'$\bigstar$ heating$ -$')
ax1.plot(pm1.r, pm1.density['NFW']['SH'], c = 'cyan', label = r'$\bigstar$ heating')

ax2.plot(pm1.r, pm1.density['Core']['GS'], c = 'b', label = 'GS')
ax2.plot(pm1.r, pm1.density['Core']['no spike'], c = 'k', label = 'no spike')
ax2.plot(pm1.r, pm1.density['Core']['NA'], c = 'm', label = 'NA')
ax2.plot(pm1.r, pm1.density['Core']['BM'], c = 'grey', label = 'BM')
ax2.plot(pm1.r, pm1.density['Core']['SH-'], c = 'brown', label = r'$\bigstar$ heating$ -$')
ax2.plot(pm1.r, pm1.density['Core']['SH'], c = 'cyan', label = r'$\bigstar$ heating')

ax1.set_xscale('log')
ax1.set_yscale('log')
xticks = [10**i for i in range(-9,1)]
ax1.set_xticks(xticks)
ax1.set_xlim(8e-10,1.2e0)
yticks = [10**i for i in range(0,19,2)]
ax1.set_yticks(yticks)
ax1.set_ylim(1e0,5e18)
ax1.tick_params(which = 'both', direction = 'in', top = True, right = True)
ax1.legend(framealpha = 0)
ax1.grid()


ax2.set_xscale('log')
ax2.set_yscale('log')
xticks = [10**i for i in range(-9,1)]
ax2.set_xticks(xticks)
ax2.set_xlim(8e-10,1.2e0)
yticks = [10**i for i in range(0,19,2)]
ax2.set_yticks(yticks)
ax2.set_ylim(1e0,5e16)
ax2.tick_params(which = 'both', direction = 'in', top = True, right = True)
ax2.legend(framealpha = 0)
ax2.grid()
```
